```python
import requests
import bs4

url = "https://www.ptt.cc/bbs/Food/index.html"

responses = requests.get(url)
soup = bs4.BeautifulSoup(responses.text, "lxml")

# articles = soup.find_all("div", {"class":, "r-ent"})
articles = soup.find_all("div", "r-ent")
# articles = [ r-ent soup element, r-ent soup element, r-ent soup element,]


for article in articles:
    meta = article.find("div", "title").find("a")
    if meta == None:
        continue

    title = meta.getText().strip()

    if "食記" not in title:
        continue

    link = meta.get("href")
    push = article.find("div", "nrec").getText()
    if len(push) == 0:
        push = 0
    else:
        push = int(push)

    date = article.find("div", "date").getText()
    author = article.find("div", "author").getText()

    print(push, title, date, author, link)
```

```python
import requests
import bs4
import pickle

url = "https://www.ptt.cc/bbs/Food/index.html"

def get_posts_on_page(this_url):
    responses = requests.get(this_url)
    soup = bs4.BeautifulSoup(responses.text, "lxml")


    articles = soup.find_all("div", "r-ent")

    posts = list()

    for article in articles:
        meta = article.find("div", "title").find("a")
        if meta == None:
            continue

        title = meta.getText().strip()

        if "食記" not in title:
            continue

        link = meta.get("href")
        push = article.find("div", "nrec").getText()
        if len(push) == 0:
            push = 0
        else:
            push = int(push)

        date = article.find("div", "date").getText()
        author = article.find("div", "author").getText()

        posts.append({"push":push, "title":title, "date":date, "author":author, "link":link})

        next_link = soup.find("div", "btn-group-paging").find_all("a", "btn")[1].get("href")

    return posts, next_link

def get_pages(num):
    page_url = url

    all_posts = list()
    for i in range(num):
        posts, link = get_posts_on_page(page_url)
        all_posts += posts

        page_url = urllib.parse.urljoin(url, link)
        print("--> crawled page:", i)
    return all_posts

num = 10
posts = get_pages(num)
with open("food_post.pkl", "wb") as f:
# f = open("food_post.pkl", "wb")
    pickle.dump(posts, f)
# f.close()
```

```python
import pickle
import urllib.parse
import requests
import time

url = "https://www.ptt.cc/bbs/Food/index.html"

def updtae_article(post):
    post_url = urllib.parse.urljoin(url, post["link"])
    response = requests.get(post_url)
    post["content"] = response.text

posts = list()
with open("food_post.pkl", "rb") as f:
    posts = pickle.load(f)

start = time.time()
for i in range(len(posts)):
    updtae_article(posts[i])
    print("----> crawled:", i, "/", len(posts))

print("Finished, spend time:", time.time() - start, "seconds")

```

# 5.1 計算字數

```python
import pickle

posts = list()
with open("food_articles.pkl", "rb") as f:
    posts = pickle.load(f)

sum(len(post["content"]) for post in posts)
```

```python
# 字數最多

max_post_len = -1
max_post_link = ""
max_post_title = ""
max_post_content = ""

for post in posts:
    post_len = len(post["content"])
    if post_len > max_post_len:
        max_post_len = post_len
        max_post_link = post["link"]
        max_post_title = post["title"]
        max_post_content = post["content"]

print(max_post_len, max_post_link, max_post_title, max_post_content)
```

```python
# 字數最少

import sys

min_post_len = sys.maxsize
min_post_link = ""
min_post_title = ""
min_post_content = ""

for post in posts:
    post_len = len(post["content"])
    if post_len < min_post_len:
        min_post_len = post_len
        min_post_link = post["link"]
        min_post_title = post["title"]
        min_post_content = post["content"]

print(min_post_len, min_post_link, min_post_title, min_post_content)
```

# 5.2 尋找推文最多



```python
# 字數最多

max_post_push = -1
max_post_link = ""
max_post_title = ""
max_post_content = ""

for post in posts:
    post_push = post["push"]
    if post_push > max_post_push:
        max_post_push = post_push
        max_post_link = post["link"]
        max_post_title = post["title"]

print(max_post_push, max_post_link, max_post_title)
```

# 5.3 尋找關鍵字


```python
# 單一關鍵字

keyword = "台北"
searched_posts = list()

for post in posts:
    if keyword in post["title"] or keyword in post["content"]:
        searched_posts.append((post["link"], post['title'])) 
print(searched_posts)
```

```python
# 多個關鍵字

keywords = ["台北", "火鍋", "和牛"]
searched_posts = list()

def keywords_in_post(keywords, post):
    # [true, false], [true, true]
    # all: 全部list中的bool要true 才回傳true
    # any: 任一list中的bool是true 就回傳true
    return all(keyword in post['title'] or keyword in post["content"] for keyword in keywords)


for post in posts:
    if keywords_in_post(keywords, post):
        searched_posts.append((post["link"], post['title']))
        
print(searched_posts)
```
